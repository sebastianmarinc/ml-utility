#30 items, 3 traits, 500

```r
items = responses_ls[[1]]
q = q_matrix_ls[[1]]
```

```r
paste0(
  "F1 = ", paste(which(q[1,] == 1), collapse = ","), "\n",
  "F2 = ", paste(which(q[2,] == 1), collapse = ","), "\n",
  "F3 = ", paste(which(q[3,] == 1), collapse = ","), "\n",
  "COV=F1*F2*F3"
) -> spec
```

```r
system.time({ mirt_em <- mirt(items, spec, "2PL", method = "EM", SE = T, technical = list(NCYCLES = 1000)) })
```

```r
M2(mirt_em)
```

```r
data <- as.matrix(items)
Q <- as.matrix(q)
num_items <- as.double(dim(Q)[2])
num_skills <- as.double(dim(Q)[1])
num_students <- dim(data)[1]
means <- rep(0,num_skills)
enc_arch <- c(16L,8L)
enc_act <- c('relu','tanh')
out_act <- 'sigmoid'
kl <- 1
cov <- as.matrix(coef(mirt_em,simplify=T)$cov)
models_cor <- build_vae_correlated(num_items,
num_skills,
Q,
model_type = 2,
mean_vector = means,
covariance_matrix = cov,
enc_hid_arch = enc_arch,
hid_enc_activations = enc_act,
output_activation = out_act,
kl_weight = kl)
encoder_cor <- models_cor[[1]]
decoder_cor <- models_cor[[2]]
vae_cor <- models_cor[[3]]
num_train <- floor(0.8 * num_students)
num_test <- num_students - num_train
data_train <- data[1:num_train,]
data_test <- data[(num_train+1):num_students,]
num_epochs <- 96
batch_size <- 16
```

```r
system.time({
  history_cor <- train_model(vae_cor, data, num_epochs = num_epochs, batch_size = batch_size) 
})
```

```r
plot(history_cor, main = 'Training Loss',sub = '',xlab = 'Epoch',ylab = 'Loss')
```

```r
item_param_estimates_cor <- get_item_parameter_estimates(decoder_cor, model_type = 2)
diff_est_cor <- item_param_estimates_cor[[1]]
disc_est_cor <- item_param_estimates_cor[[2]]
all_theta_est_cor <- get_ability_parameter_estimates(encoder_cor, data)[[1]]
```

```r
bind_cols(diff_est_cor = diff_est_cor, diff_mirt = coef(mirt_em,simplify=T)$item[,4]) %>% 
ggplot() + aes(diff_est_cor, diff_mirt) + geom_point()
```

```r
bind_cols(diff_est_cor = diff_est_cor, diff_mirt = coef(mirt_em,simplify=T)$item[,4]) %>% cor()
```

```r
bind_cols(disc_est_cor = disc_est_cor[1,], diff_mirt = coef(mirt_em,simplify=T)$item[,1]) %>% cor()
```

```r
bind_cols(disc_est_cor = disc_est_cor[2,], diff_mirt = coef(mirt_em,simplify=T)$item[,2]) %>% cor()
```

```r
bind_cols(disc_est_cor = disc_est_cor[3,], diff_mirt = coef(mirt_em,simplify=T)$item[,3]) %>% cor()
```

```r
cor(cbind(all_theta_est_cor, fscores(mirt_em))) %>% round(.,3)
```
